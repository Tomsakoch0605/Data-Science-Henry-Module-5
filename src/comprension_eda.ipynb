```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Verificar la columna tendencia_ingresos que tiene valores numéricos mezclados
print("\nValores únicos en 'tendencia_ingresos':")
print(df['tendencia_ingresos'].value_counts())

# Identificar tipos de datos inconsistentes
print("\nRegistros con valores numéricos en 'tendencia_ingresos':")
tendencia_numerica = pd.to_numeric(df['tendencia_ingresos'], errors='coerce')
registros_problematicos = df[tendencia_numerica.notna() & df['tendencia_ingresos'].notna()]
print(f"Total de registros con valores numéricos: {len(registros_problematicos)}")
```

```python
duplicados = df.duplicated().sum()
print(f"\nTotal de registros duplicados: {duplicados}")

if duplicados > 0:
    print("\nPrimeros registros duplicados:")
    print(df[df.duplicated(keep=False)].head())
```

```python
# Variables numéricas clave
vars_numericas = ['capital_prestado', 'plazo_meses', 'edad_cliente', 'salario_cliente', 
                  'cuota_pactada', 'puntaje', 'puntaje_datacredito']

print("\n3.1 Análisis de Asimetría (Skewness) y Curtosis:")
print("-" * 60)
for var in vars_numericas:
    skew = df[var].skew()
    kurt = df[var].kurtosis()
    print(f"{var:30s} | Skewness: {skew:7.2f} | Kurtosis: {kurt:7.2f}")
```

```python
def detectar_outliers_iqr(df, columna):
    Q1 = df[columna].quantile(0.25)
    Q3 = df[columna].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[columna] < lower_bound) | (df[columna] > upper_bound)]
    return len(outliers), (lower_bound, upper_bound)

print("\n4.1 Outliers por método IQR:")
print("-" * 60)
for var in vars_numericas:
    n_outliers, bounds = detectar_outliers_iqr(df, var)
    pct = (n_outliers / len(df)) * 100
    print(f"{var:30s} | Outliers: {n_outliers:5d} ({pct:5.2f}%)")
```

```python
print("\nDistribución de la variable objetivo:")
print(df['Pago_atiempo'].value_counts())
print("\nPorcentaje:")
print(df['Pago_atiempo'].value_counts(normalize=True) * 100)

# Balance de clases
balance = df['Pago_atiempo'].value_counts()
ratio = balance.min() / balance.max()
print(f"\nRatio de balance de clases: {ratio:.4f}")
if ratio < 0.3:
    print("⚠️ ADVERTENCIA: Dataset desbalanceado detectado")
```

```python
# Correlaciones con Pago_atiempo
correlaciones = df[vars_numericas + ['Pago_atiempo']].corr()['Pago_atiempo'].sort_values(ascending=False)
print("\n6.1 Correlaciones con 'Pago_atiempo':")
print(correlaciones)
```

```python
# tipo_laboral
print("\nDistribución por tipo_laboral:")
print(df['tipo_laboral'].value_counts())
print("\nTasa de pago a tiempo por tipo_laboral:")
print(df.groupby('tipo_laboral')['Pago_atiempo'].agg(['mean', 'count']))

# tipo_credito
print("\nDistribución por tipo_credito:")
print(df['tipo_credito'].value_counts().sort_index())
print("\nTasa de pago a tiempo por tipo_credito:")
print(df.groupby('tipo_credito')['Pago_atiempo'].agg(['mean', 'count']))
```

```python
# Crear ratios
df_temp = df.copy()
df_temp['ratio_cuota_salario'] = df_temp['cuota_pactada'] / df_temp['salario_cliente']
df_temp['ratio_deuda_ingreso'] = df_temp['total_otros_prestamos'] / df_temp['salario_cliente']
df_temp['ratio_prestamo_salario'] = df_temp['capital_prestado'] / df_temp['salario_cliente']

print("\n8.1 Estadísticas de ratios financieros:")
ratios = ['ratio_cuota_salario', 'ratio_deuda_ingreso', 'ratio_prestamo_salario']
for ratio in ratios:
    print(f"\n{ratio}:")
    print(df_temp[ratio].describe())

# Correlación de ratios con pago a tiempo
print("\n8.2 Correlación de ratios con Pago_atiempo:")
for ratio in ratios:
    corr = df_temp[[ratio, 'Pago_atiempo']].corr().iloc[0, 1]
    print(f"{ratio:30s}: {corr:.4f}")
```

```python
df_temp['año_prestamo'] = df['fecha_prestamo'].dt.year
df_temp['mes_prestamo'] = df['fecha_prestamo'].dt.month
df_temp['trimestre_prestamo'] = df['fecha_prestamo'].dt.quarter

print("\n9.1 Distribución de préstamos por año:")
print(df_temp['año_prestamo'].value_counts().sort_index())

print("\n9.2 Distribución de préstamos por mes:")
print(df_temp['mes_prestamo'].value_counts().sort_index())

print("\n9.3 Tasa de pago a tiempo por año:")
print(df_temp.groupby('año_prestamo')['Pago_atiempo'].mean())
```
